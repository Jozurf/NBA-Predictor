### Make rolling averages

```python
non_rolling_cols = player_df.columns[~player_df.columns.str.contains('_')] 
```

```python
non_rolling_cols
```

```python
home_rolling_cols = [col for col in list(player_df.columns) if col not in non_rolling_cols and "_opp" not in col]
```

```python
home_rolling_cols
```

```python
window_size = 82 
df_home_rolling = player_df[home_rolling_cols + ["team", "season"]]
def find_team_averages(team):
    rolling = team[home_rolling_cols].shift(1).rolling(window_size, min_periods=1).mean()
    return rolling
    #using rolling average on team
df_home_rolling = df_home_rolling.groupby(["team", "season"], group_keys=False).apply(find_team_averages)
df_home_rolling = df_home_rolling.add_suffix(f'_{window_size}')
```

```python
df_home_rolling
```

```python
away_rolling_cols = [col for col in list(player_df.columns) if col not in non_rolling_cols and "_opp" in col and col != "team_opp"]
```

```python
window_size = 82 
df_away_rolling = player_df[away_rolling_cols + ["team_opp", "season"]]
def find_team_averages(team):
    rolling = team[away_rolling_cols].shift(1).rolling(window_size, min_periods=1).mean()
    return rolling
    #using rolling average on team
df_away_rolling = df_away_rolling.groupby(["team_opp", "season"], group_keys=False).apply(find_team_averages)
df_away_rolling = df_away_rolling.add_suffix(f'_{window_size}')
```

```python
player_df = pd.concat([player_df, df_home_rolling, df_away_rolling], axis=1)
```

```python
player_df
```

```python
player_rolling_cols = [s + "_82" for s in home_rolling_cols] + [s + "_82" for s in away_rolling_cols]
```

```python
player_rolling_cols
```

```python
features_to_train = player_rolling_cols
```

```python
# Extract the column names for home and away players
extra = ["team", "team_opp", "date", "season", "won"]
away_players_columns = [col for col in processed_pdf.columns if col.endswith('opp_82')]
home_players_columns = [col for col in processed_pdf.columns if 'opp_82' not in col and col not in extra]
```

```python
# Extract the unique players
players = pd.unique([col.split('_')[1] for col in processed_pdf.columns if '_' in col and 'opp' not in col])
```

```python
# Extract the features columns
features = pd.unique([col.split('_')[0] for col in processed_pdf.columns if '_' in col and 'team' not in col])
```

```python
processed_pdf=processed_pdf.reset_index(drop="index")
```

```python
processed_pdf.to_csv("player_df_processed.csv")
```

```python
import numpy as np

# Create an empty 4D matrix
player_data_matrix = np.zeros((len(processed_pdf), len(players), len(features), 2))  # 2 for home and away

# Iterate over each game
for i, game_row in processed_pdf.iterrows():
    # Iterate over home and away teams
    for j, team in enumerate(['', '_opp']):
        # Iterate over players
        for k, player in enumerate(players):
            # Get the player's data
            player_data = game_row[[f'{feature}_{player}{team}_82' for feature in features]].values
            # Assign the player's data to the 4D matrix
            player_data_matrix[i, k, :, j] = player_data
```
